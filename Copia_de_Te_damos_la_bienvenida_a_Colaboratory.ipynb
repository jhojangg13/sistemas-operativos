```
from graphviz import Digraph

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, value):
        if self.root is None:
            self.root = Node(value)
        else:
            self._insert(self.root, value)

    def _insert(self, current, value):
        if value < current.value:
            if current.left is None:
                current.left = Node(value)
            else:
                self._insert(current.left, value)
        elif value > current.value:
            if current.right is None:
                current.right = Node(value)
            else:
                self._insert(current.right, value)

    def visualize(self):
        if not self.root:
            return "El árbol está vacío"
        graph = Digraph(format='png')
        self._add_nodes_edges(graph, self.root)
        return graph

    def _add_nodes_edges(self, graph, node):
        if node is None:
            return
        graph.node(str(node.value), str(node.value))
        if node.left:
            graph.edge(str(node.value), str(node.left.value))
            self._add_nodes_edges(graph, node.left)
        if node.right:
            graph.edge(str(node.value), str(node.right.value))
            self._add_nodes_edges(graph, node.right)

def create_and_visualize_bst(values):
    bst = BinarySearchTree()
    for value in values:
        bst.insert(value)
    graph = bst.visualize()
    graph.render('binary_search_tree', format='png', cleanup=True)
    return graph

inputs = {
    "A": [15, 6, 18, 3, 7, 17, 20, 2, 4, 13, 9],
    "B": [25, 15, 50, 10, 22, 35, 70, 4, 12, 18, 24, 31, 44, 66, 90],
    "C": [30, 20, 40, 10, 25, 35, 50],
    "D": [27, 14, 35, 10, 19, 31, 42, 9, 12, 18, 20, 30, 33, 38, 45]
}

for key, values in inputs.items():
    print(f"Árbol {key}")
    graph = create_and_visualize_bst(values)
    display(graph)
```
